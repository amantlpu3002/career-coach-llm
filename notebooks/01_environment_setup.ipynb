## 1.4 · Verify CUDA is available

```python
import torch

print(f"PyTorch version : {torch.__version__}")
print(f"CUDA available  : {torch.cuda.is_available()}")

if torch.cuda.is_available():
    name  = torch.cuda.get_device_name(0)
    total = torch.cuda.get_device_properties(0).total_memory / 1e9
    print(f"GPU             : {name}")
    print(f"VRAM            : {total:.1f} GB")
    if total < 4.0:
        print("⚠️  Under 4 GB VRAM — set batch_size=2 in notebook 03.")
    else:
        print("✅ Good to go for QLoRA fine-tuning!")
else:
    print("❌ CUDA not available. Reinstall PyTorch with the correct CUDA version.")
```
